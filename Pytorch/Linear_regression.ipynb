# **Simple Linear Regression**

one input data, one output data

```
import torch
```

```
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[2],[4],[6]])
```

```
w = torch.zeros(1, requires_grad=True) #requires_grad=True : This data will learn
b = torch.zeros(1,requires_grad=True)
hypothesis = x_train * w + b
```

```
cost = torch.mean((hypothesis - y_train) ** 2)
```

```
optimizer = torch.optim.SGD([w,b], lr = 0.01) #lr : learning rate

optimizer.zero_grad() #init gradient
cost.backward()  #calc gradient
optimizer.step()  #improve
```

```
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[2],[4],[6]])

w = torch.zeros(1, requires_grad=True) #requires_grad=True : This data will learn
b = torch.zeros(1,requires_grad=True)

optimizer = torch.optim.SGD([w,b], lr = 0.01) #lr : learning rate

nb_epochs = 1000
for epoch in range(1, nb_epochs + 1):
  hypothesis = x_train * w + b
  cost = cost = torch.mean((hypothesis - y_train) ** 2)

  optimizer.zero_grad() #init gradient to 0
  cost.backward()  #calc gradient
  optimizer.step()  #gradient descent

  print('w: ',w.item())
  print('b: ',b.item())
```

# **Multivariate Linear Regression**

many input data, one output data

```
x_train = torch.FloatTensor([[73,80,75],
                             [93,88,93],
                             [89,91,90],
                             [96,98,100],
                             [73,66,70]])
y_train = torch.FloatTensor([[152],[185],[180],[196],[142]])
```

```
w = torch.zeros((3,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```
hypothesis = x_train.matmul(w) + b   # w1 * x1_train + w2 * x2_train + w3 * x3_train + b
```

```
x_train = torch.FloatTensor([[73,80,75],
                             [93,88,93],
                             [89,91,90],
                             [96,98,100],
                             [73,66,70]])
y_train = torch.FloatTensor([[152],[185],[180],[196],[142]])

w = torch.zeros((3,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = torch.optim.SGD([w,b], lr = 1e-5) #lr : learning rate

nb_epochs = 20
for epoch in range(1, nb_epochs + 1):
  hypothesis = x_train.matmul(w) + b   # w1 * x1_train + w2 * x2_train + w3 * x3_train + b

  cost = torch.mean((hypothesis - y_train)**2)

  optimizer.zero_grad() #init gradient to 0
  cost.backward()  #calc gradient
  optimizer.step()  #gradient descent

  print('w: ',hypothesis.squeeze().detach())
  print('b: ',cost.item())
```

```
# 데이터
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
# 모델 초기화
model = MultivariateLinearRegressionModel()
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr=1e-5)

nb_epochs = 20
for epoch in range(nb_epochs+1):
    
    # H(x) 계산
    prediction = model(x_train)
    
    # cost 계산
    cost = F.mse_loss(prediction, y_train)
    
    # cost로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    # 20번마다 로그 출력
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```
