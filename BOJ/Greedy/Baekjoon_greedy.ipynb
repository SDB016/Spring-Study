# **1439**

image.png

```
import sys

data = sys.stdin.readline()
p = []
cnt = 0

for i in range(len(data) - 2):
    if data[i] != data[i+1]:
        p.append(i + 1)
cnt = len(p)
if cnt%2 == 0:
    cnt //= 2
else:
    cnt = (cnt+1)//2

print(cnt)
```

다른 풀이

```
data = input()
cnt = 0

for i in range(len(data) - 1):
    if data[i] != data[i+1]:
        cnt += 1

print((cnt+1)//2)
```

# **1541** (틀렸던 문제)

image.png

```
data = input().split('-')
res = 0
for i in data[0].split('+'):
    res+=int(i)
for i in data[1:]:
    for j in i.split('+'):
        res -= int(j)
print(res)
```

# **4796**

image.png

```
result = []
while True:
    l,p,v = map(int,input().split())
    if l==0 and p==0 and v==0:
        break
    result.append((v//p)*l+min(l,v%p))
for i in range(len(result)):
    print('Case {}: {}'.format(i+1,result[i]))
```

# **2810**

image.png

```
n = int(input())
seats = input()
if n != 1:
    sCnt = seats.count('S')
    res = sCnt+(n - sCnt)//2
    if 'LL' in seats:
        res += 1
    print(res)
else:
    print(1)
```

다른 코드

```
n = int(input())
seats = input()
cnt = seats.count('LL')
if (cnt <= 1):
    print(len(seats))
else:
    print(len(seats)- cnt + 1)
```
