```
import sys

data = sys.stdin.readline()
p = []
cnt = 0

for i in range(len(data) - 2):
    if data[i] != data[i+1]:
        p.append(i + 1)
cnt = len(p)
if cnt%2 == 0:
    cnt //= 2
else:
    cnt = (cnt+1)//2

print(cnt)
```

# **17609** (틀린 문제, 어려움)

image.png

```
import sys
def secCheck(word,left,right):
    while(left<right):
        if(word[left]==word[right]):
            left += 1
            right -= 1
        else:
            return False
    return True
def firstCheck(word,left,right):
    while(left<right):
        if(word[left]==word[right]):
            left += 1
            right -= 1
        else:
            check1 = secCheck(word,left+1,right)
            check2 = secCheck(word,left,right-1)
            if(check1 or check2):
                return 1
            else:
                return 2
    return 
n = int(sys.stdin.readline().rstrip("\n"))
for _ in range(n):
    word = sys.stdin.readline().rstrip("\n")
    left, right = 0, len(word)-1
    print(firstCheck(word,left,right))
```

# **1744** (틀린 문제, 다소 어려움)

image.png

```
import sys
minus = []
plus = []
one = 0
def cal(arr):
    res = 0
    for i in range(0,len(arr)-1,2):
        res += arr[i] * arr[i+1]
    if len(arr)%2 != 0:
        res += arr[-1]
    return res
for _ in range(int(sys.stdin.readline().rstrip("\n"))):
    num = int(sys.stdin.readline().rstrip("\n"))
    if num > 1:
        plus.append(num)
    elif num <= 0:
        minus.append(num)
    elif num == 1:
        one += 1
plus.sort(reverse = True)
minus.sort()
print(cal(plus)+cal(minus)+one)
```

# **1946** (다소 어려움)

image.png

```
import sys
input = sys.stdin.readline
for _ in range(int(input())):
    n= int(input())
    grades = [list(map(int, input().split())) for _ in range(n)]
    grades.sort(key = lambda x:x[0])
    interview = [grades[i][1] for i in range(n)]
    min = interview[0]
    res = 1
    for i in interview:
        if min > i:
            res += 1
            min = i
    print(res)
```

# **1339** (어려움)

image.png

```
operands = [input() for _ in range(int(input()))]   
alpha = [0 for _ in range(26)]
for operand in operands:
    for i in range(len(operand)):
        alpha[ord(operand[i])-ord('A')] += 10**(len(operand)-i-1)
alpha.sort(reverse = True)
res, bias = 0, 9
for i in alpha:
    res += i * bias
    bias -= 1
print(res)
```
