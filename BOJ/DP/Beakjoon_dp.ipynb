```
import sys
input = sys.stdin.readline

for _ in range(int(input())):
    n = int(input())
    dp = [1] * 12
    dp[0], dp[1], dp[2], dp[3] = 0, 1, 2, 4
    for i in range(4, n+1):
        dp[i] = dp[i-1]+dp[i-2]+dp[i-3]
    print(dp[n])
```

```
import sys
input = sys.stdin.readline
n = int(input())

dp = [0]*1000001

for i in range(2,n+1):
    dp[i] = dp[i-1] + 1

    if i % 3 == 0:
        dp[i] = min(dp[i],dp[i//3]+1)
    if i % 2 == 0:
        dp[i] = min(dp[i],dp[i//2]+1)
    
print(dp[n])
```

```
import sys
input = sys.stdin.readline
n = int(input())
stairs = [0] * 301
dp = [0] * 301
for i in range(n):
    stairs[i] = int(input())
dp[0], dp[1], dp[2] = stairs[0], stairs[0] + stairs[1], max(stairs[0] + stairs[2], stairs[1] + stairs[2])

for i in range(3, n):
    dp[i] = max(stairs[i] + dp[i - 2], stairs[i] + dp[i - 3] + stairs[i - 1])
print(dp[n-1])
```
