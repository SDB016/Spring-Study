```
import sys
input = sys.stdin.readline

for _ in range(int(input())):
    n = int(input())
    dp = [1] * 12
    dp[0], dp[1], dp[2], dp[3] = 0, 1, 2, 4
    for i in range(4, n+1):
        dp[i] = dp[i-1]+dp[i-2]+dp[i-3]
    print(dp[n])
```
