```
import sys
input = sys.stdin.readline

for _ in range(int(input())):
    n = int(input())
    dp = [1] * 12
    dp[0], dp[1], dp[2], dp[3] = 0, 1, 2, 4
    for i in range(4, n+1):
        dp[i] = dp[i-1]+dp[i-2]+dp[i-3]
    print(dp[n])
```

```
import sys
input = sys.stdin.readline
n = int(input())

dp = [0]*1000001

for i in range(2,n+1):
    dp[i] = dp[i-1] + 1

    if i % 3 == 0:
        dp[i] = min(dp[i],dp[i//3]+1)
    if i % 2 == 0:
        dp[i] = min(dp[i],dp[i//2]+1)
    
print(dp[n])
```

```
import sys
input = sys.stdin.readline
n = int(input())
stairs = [0] * 301
dp = [0] * 301
for i in range(n):
    stairs[i] = int(input())
dp[0], dp[1], dp[2] = stairs[0], stairs[0] + stairs[1], max(stairs[0] + stairs[2], stairs[1] + stairs[2])

for i in range(3, n):
    dp[i] = max(stairs[i] + dp[i - 2], stairs[i] + dp[i - 3] + stairs[i - 1])
print(dp[n-1])
```

```
import sys
input = sys.stdin.readline

n = int(input())
triangles = []
for i in range(n):
    triangles.append(list(map(int, input().split())))
    for j in range(n-len(triangles[i])):
        triangles[i].append(0)
dp = [[0]*501 for _ in range(501)]
dp[0] = triangles[0]
for i in range(1,n):
    for j in range(0,i+1):
        if j==0:
            dp[i][j] = dp[i-1][j]+triangles[i][j]
        elif j==i:
            dp[i][j] = dp[i-1][j-1]+triangles[i][j]
        else:
            dp[i][j] = max(dp[i-1][j-1], dp[i-1][j]) + triangles[i][j]
print(max(dp[n-1]))
```

```
import sys
input = sys.stdin.readline

n = int(input())
glasses = [0]
for _ in range(n):
    glasses.append(int(input()))
dp = [0]
dp.append(glasses[1])
if n>1:
    dp.append(glasses[1]+glasses[2])
for i in range(3, n+1):
    dp.append(max(dp[i-3]+glasses[i-1]+glasses[i], dp[i-2] + glasses[i], dp[i-1]))
print(dp[n])
```
