```
import sys
input = sys.stdin.readline
cnt = 0
for _ in range(int(input())):
    word = input()
    cnt += list(word) == sorted(word, key = word.find)
print(cnt)
```

`sorted(word, key = word.find)` 로 정렬하면 찾아지는 character순서대로 정렬이 된다.

```
import sys
input = sys.stdin.readline
n = int(input())
people = []
res = []
for _ in range(n):
    people.append(list(map(int, input().split())))

for current in people:
    rank = 1
    for nextp in people:
        if current != nextp:
            if current[0] < nextp[0] and current[1] < nextp[1]:
                rank += 1
    print(rank)
     
```

```
import sys
input = sys.stdin.readline
for _ in range(int(input())):
    n, m = map(int, input().split())
    imp = list(map(int, input().split()))
    idx = list(range(len(imp)))
    idx[m] = 'target'
    count = 0
    while True:
        if imp[0] == max(imp):
            count += 1
            if idx[0] == 'target':
                print(count)
                break
            imp.pop(0)
            idx.pop(0)
        else:
            imp.append(imp.pop(0))
            idx.append(idx.pop(0))        
```

```
import sys
input = sys.stdin.readline

dy = [0,1,0,-1]
dx = [-1,0,1,0] #북동남서

def clean(x, y, d):
    global res
    if maps[x][y] == 0:
        maps[x][y] = 2
        res += 1
    for _ in range(4):
        nd = (d + 3) % 4 #왼쪽 방향
        nx = x + dx[nd]
        ny = y + dy[nd]
        if maps[nx][ny] == 0:
            clean(nx, ny, nd)
            return
        d = nd
    nd = (d + 2) % 4
    nx = x + dx[nd]
    ny = y + dy[nd]
    if maps[nx][ny] == 1:
        return
    clean(nx, ny, d)

n, m = map(int, input().split())
x, y, d = map(int, input().split())
maps = [list(map(int, input().split())) for _ in range(n)]

res = 0
clean(x, y, d)
print(res)
```
