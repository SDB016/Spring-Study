# **DFS 소스코드**

```
def dfs(graph, v, visited):
    visited[v] = True
    print(v, end=' ')
    
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)

graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * (len(graph))
dfs(graph, 1, visited)
```

# **BFS 소스코드**

```
from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True

    while queue:
        v = queue.popleft()
        print(v,end=' ')

        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * (len(graph))
bfs(graph, 1, visited)
```

```
import sys
from collections import deque
input = sys.stdin.readline

n = int(input())
links = [[] for _ in range(n+1)]
visited = [False] * (n+1)

for _ in range(int(input())):
    a,b = map(int, input().split())
    links[a].append(b)
    links[b].append(a)

def bfs(v):
    queue = deque([v])
    visited[v] = True

    while queue:
        node = queue.popleft()
        for i in links[node]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True

bfs(1)
print(visited.count(True)-1)
```

```
import sys
from collections import deque
input = sys.stdin.readline

n, m = map(int, input().split())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().rstrip())))
move = [[-1,0],[1,0],[0,1],[0,-1]]
visited = [[False] * m for _ in range(n)]

def bfs(x,y):
    queue = deque([(x,y)])
    visited[x][y] = True
    while queue:
        x,y = queue.popleft()
        for i in range(4):
            nx = x+move[i][0]
            ny = y+move[i][1]
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] == 0:
                continue
            if not visited[nx][ny]:
                visited[nx][ny] = True
                queue.append((nx,ny))
                graph[nx][ny] = graph[x][y] + 1

bfs(0,0)
print(graph[n-1][m-1])
```

```
import sys
from collections import deque
input = sys.stdin.readline

n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().rstrip())))
visited = [[False] * n for _ in range(n)]
move = [[0,1],[0,-1],[1,0],[-1,0]]
res = []

def bfs(x,y):
    if graph[x][y] == 0 or visited[x][y]:
        return False
    cnt = 1
    queue = deque([(x,y)])
    visited[x][y] = True
    while queue:
        x,y = queue.popleft()
        for d in move:
            nx, ny = x + d[0], y + d[1]
            if nx < 0 or ny < 0 or nx >= n or ny >= n:
                continue
            if graph[nx][ny] == 0:
                continue
            if not visited[nx][ny]:
                queue.append((nx, ny))
                visited[nx][ny] = True
                cnt += 1
    res.append(cnt)


for i in range(n):
    for j in range(n):
        bfs(i,j)
print(len(res))
for i in sorted(res):
    print(i)
```

```
import sys
from collections import deque
input = sys.stdin.readline

n, k = map(int, input().split())
visited = [0] * 100001
queue = deque()
queue.append([n,0])

while queue:
    v, depth = queue[0][0], queue[0][1]
    if v == k:
        break
    queue.popleft()
    visited[v] = 1
    if v*2 <= 100000 and visited[v*2] == 0:   
        queue.append([v*2, depth+1])
    if v+1 <= 100000 and visited[v+1] == 0:
        queue.append([v+1, depth+1])
    if v-1 >= 0 and visited[v-1] == 0:
        queue.append([v-1, depth+1])

print(queue[0][1])
```

```
import sys
from collections import deque
input = sys.stdin.readline


for _ in range(int(input())):
    m,n,k = map(int, input().split())
    move = [[0,1],[0,-1],[1,0],[-1,0]]
    graph = [[0]*m for _ in range(n)]
    visit = [[0]*m for _ in range(n)]
    cnt = 0
    for _ in range(k):
        a,b = map(int, input().split())
        graph[b][a] = 1
    
    def bfs(x,y):
        global cnt
        if visit[x][y] == 0 and graph[x][y] == 1:
            cnt += 1
            queue = deque([(x,y)])
            while queue:
                x, y = queue.popleft()
                for i in range(4):
                    nx, ny = x+move[i][0], y+move[i][1]
                    if 0 <= nx < n and 0 <= ny < m and visit[nx][ny] == 0 and graph[nx][ny] == 1:
                        visit[nx][ny] = 1
                        graph[nx][ny] = 2
                        queue.append((nx,ny))
    
    for i in range(n):
        for j in range(m):
            bfs(i,j)
    print(cnt)
```
